```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
import numpy as np
import pandas as pd
```

```python
#Import the data
df = pd.read_csv('/content/drive/MyDrive/Case Studies/JP Morgan&Chase/Task 3/Loan_Data.csv')
df
```

```python
#Creating debt_to_income ratio
df['debt_to_income'] = df['total_debt_outstanding']/df['income']
```

```python
#Creating payment_to_income ratio
df['payment_to_income'] = df['loan_amt_outstanding']/df['income']
```

```python
#Define variable features
features = ['credit_lines_outstanding','debt_to_income','payment_to_income','years_employed','fico_score']
```

```python
#Using a logistic regression model
model = LogisticRegression(solver='liblinear',tol=1e-5,max_iter=1000)
model.fit(df[features],df['default'])

print(model.coef_ , model.intercept_)
```

```python
#Making predictions
y_pred = model.predict(df[features])

fpr, tpr, thresholds = metrics.roc_curve(df['default'],y_pred)

print((1.0 * (abs(df['default'] - y_pred)).sum()) / len(df))
print(metrics.auc(fpr,tpr))
```
