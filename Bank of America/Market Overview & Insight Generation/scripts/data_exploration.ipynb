```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
excel_file_path = '/content/drive/MyDrive/Case Studies/Bank of America/Sample Dataset - Task 1.xlsx'
```

```python
df = pd.read_excel(excel_file_path,sheet_name="Sheet1",header=1)
```

### How has Interest rate changed with time?

```python
#Plot how interest rate changes with weeks,include legend

sns.lineplot(data=df,x="Date",y="Interest Rate (%)")
plt.title("Interest Rate Changes with Weeks")
plt.xlabel("Week Number")
plt.ylabel("Interest Rate (%)")
plt.legend(labels=["Interest Rate"])
plt.ylim(0.5,3)
plt.show()
```

* Interest rate has increased steadily over the quarter, from **1.5% to 2.0%**
* Higher interest rate is alarming for emerging sectors like technology, due to higher borrowing costs.

How has different sectors performes with time?

```python
#Plot performance of different sectors with time

plt.figure(figsize=(14,6))
sns.lineplot(data=df,x="Date",y="Technology (%)",label="Technology")
sns.lineplot(data=df,x="Date",y="Healthcare (%)",label="Healthcare")
sns.lineplot(data=df,x="Date",y="Energy (%)",label="Energy")
sns.lineplot(data=df,x="Date",y="Financials (%)",label="Financials")
plt.title(" Performance against Time")
plt.xlabel("Time")
plt.ylabel("Sectors Performance (%)")
plt.legend()
plt.show()
```

**Technology Sector Performance:**
* Early gains**(+2%, +1.5%)** were offset by declines**(-0.5%, -1%, 0.8%)** due to rate hikes and volatility.
* Late **recoveries(+0.2%, +1.0%)** were driven by positive earning reports and new givernment policies

**Healthcare Sector Performance:**
* Healthcare sectors showed consistent, moderate growth**(+0.5% to +0.8%)**, proving itself as a stable,defensive sector during volatile periods.

**Energy Sector Performance:**
* Showed initial losses **(-1.0%, -1.2%, -0.5%)** in the early quarters
* Losses were recovered after the tech earnings reports, which showed a rise in performance**(+0.5%, +1.0%, +1.5%, 2.0)**

**Financial Sector Performance:**
* Overall stable but showed slight decline**(+1.0% to -0.7%)**, possibly affected by the geopolitical tensions as well as tech earning reports.

###During which economic event was the market most volatile?

```python
#Plot Market Volatility against Economic Events

plt.figure(figsize=(14,6))
sns.barplot(data=df,x='Economic Events',y='Market Volatility (VIX)')
plt.title("Market Volatility vs Economic Events")
plt.xlabel("Economic Event")
plt.ylabel("Market Volatility (VIX)")
plt.show()
```

* Market Volatility (VIX) **peaked at 25** between **Geopolitical tensions**
  and **Tech Earnings Reports**,
* then got **stabilized towards the end of the quarter** at 20

###Which Sectors are interconnected with each other?

```python
sectors = ['Technology (%)', 'Healthcare (%)', 'Energy (%)', 'Financials (%)']
corr_matrix = df[sectors].corr()
sns.heatmap(corr_matrix,annot=True,cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

**Financials** and **Technology** sectors are highly correlated.
Implies that change in performance in innovation-driven growth highly affects the finance sectors
